```python
import requests
import matplotlib.pyplot as plt

def fetch_price_data(symbol, start_date, end_date):
    base_url = "https://api.coingecko.com/api/v3/coins/"
    url = f"{base_url}/{symbol}/market_chart/range"
    params = {
        "vs_currency": "usd",
        "from": start_date,
        "to": end_date,
    }

    response = requests.get(url, params=params)
    data = response.json()
    prices = data['prices']

    return prices

def plot_price_data(prices):
    timestamps, values = zip(*prices)
    plt.figure(figsize=(10, 6))
    plt.plot(timestamps, values, label="Price (USD)")
    plt.xlabel("Timestamp")
    plt.ylabel("Price (USD)")
    plt.title("Historical Price Data for Shiba Inu")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    symbol = "shiba-inu"
    start_date = "1622505600"  # Example: June 1, 2021
    end_date = "1625097600"    # Example: June 30, 2021

    price_data = fetch_price_data(symbol, start_date, end_date)
    plot_price_data(price_data)
```

In this example, we fetch historical price data for Shiba Inu from the CoinGecko API and plot it using matplotlib. However, keep in mind that this is just a simple example for visualization purposes and doesn't involve any predictive modeling.
